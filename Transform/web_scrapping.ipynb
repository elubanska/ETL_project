```python
from splinter import Browser
from bs4 import BeautifulSoup
import requests
from webdriver_manager.chrome import ChromeDriverManager
# How to cite: Statistics Canada. Table 18-10-0205-01 New housing price index, monthly
```

```python
url = 'https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=1810020501&pickMembers%5B0%5D=1.17&cubeTimeFrame.startMonth=01&cubeTimeFrame.startYear=2019&cubeTimeFrame.endMonth=12&cubeTimeFrame.endYear=2020&referencePeriods=20190101%2C20201201'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
title = soup.title.text
print(title)
```

```python
table = soup.find_all("table", class_ ="pub-table") 
#len(table)
#type(table)
#print(table)
```

```python
for row in table.find_all('tr'):
    for cell in row.find_all('td'):
        print(cell.text)
```
